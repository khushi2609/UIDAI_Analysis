```python
import operator 
import csv
import matplotlib.pyplot as plt
```

```python
raw=[]
data = open('../dataset.csv', 'r')
next(data)
```

```python
for line in data:
    raw.append(line.strip().split(','))
```

```python
mvk={}
for row in raw:
    k=row[2]
    v=row[8]
    if v!='':
        v=int(float(v))
    else:
        v=0
    if k in mvk:
        mvk[k]+=v
    else:
        mvk[k]=v
```

```python
sorted_d = dict( sorted(mvk.items(), key=lambda x: x[1], reverse=True))
```

```python
states=[]
Aadhaars=[]
```

```python
for key ,value in sorted_d.items():
    states.append(key)
    Aadhaars.append(value)
```

```python
for i in range(len(states)):
    print(states[i],Aadhaars[i])
```

```python
plt.plot(states,Aadhaars,marker="o")
plt.xticks(rotation=90)
plt.xlabel("States")
plt.ylabel("Number of Aadhaar Generated")
plt.savefig('States_vs_Aadhaar.png')
plt.show()
```
